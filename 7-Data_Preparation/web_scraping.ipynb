```python
from selenium import webdriver
```

```python
fawzi=webdriver.Chrome()
```

```python
my_products={"product_name":[],
             "product_price":[]}
for i in range(1,6):
    fawzi.get(f"https://www.noon.com/egypt-en/electronics-and-mobiles/mobiles-and-accessories/mobiles-20905/?isCarouselView=false&limit=200&page={i}&sort%5Bby%5D=popularity&sort%5Bdir%5D=desc")
    fawzi.implicitly_wait(30)
    products = fawzi.find_elements("xpath","//div[@class='sc-433dbb72-23 gUfohz']")
    for product in products:
        try:
            name = product.find_element("xpath", ".//div[@class='sc-433dbb72-25 elDiNI']").text.strip()
        except:
            name = "Unknown"

        try:
            price = product.find_element("xpath", ".//strong").text.strip()
        except:
            price = "N/A"
        my_products["product_name"].append(name)
        my_products["product_price"].append(price)
print(my_products)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(my_products)
df
```
