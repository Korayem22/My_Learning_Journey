```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib qt
def hamada_seq(n,c=0,draw=False):
    if n == 0:
        return None,None
    else:
        L =[0]*n
        L[0]=c
        for i in range(1,n):
            if L[i-1]-i>0:
                L[i]=L[i-1]-i
            else:
                L[i]=L[i-1]+i
        if 1 in L:
            s=L.index(1)
        else:
            s = None
        if draw:
            y_points=[]
            y_points.append(L[0])
            x=np.arange(n)
            x_points=[x[0]]
            i=1
            while i<len(L)-1:
                y_points.append(L[i])
                x_points.append(x[i])
                plt.ion()
                plt.plot(0,0)
                ax = plt.gca()           
                plt.setp(ax.lines,ydata=y_points)
                plt.setp(ax.lines,xdata=x_points)
                plt.setp(ax.lines,color='r')
                plt.xlim(0,n)
                plt.ylim(min(L)-10,max(L)+10)
                plt.pause(0.1)
                i+=1
        plt.plot(s,1,'go')
        return L,s
hamada_seq(100,10,True)
```
